# Which planets are closest to earth?


```python
earth_relation = pd.DataFrame(df, columns=['name','light_years_from_earth','planet_type', 'planet_mass', 'solar_system_name'])
earth_relation
                              
```

```python
close_to_earth = earth_relation.sort_values(by='light_years_from_earth', ascending=True)
close_to_earth.loc[~(close_to_earth==0).all(axis=1)]
close_to_earth.head(20)
```

# What are the size of these planets?

```python
sns.set_theme(style="darkgrid")

# Load the example planets dataset
planet_mass = pd.DataFrame(df, columns = ['planet_mass', 'orbital_radius','light_years_from_earth', 'discovery_date'])
planet_mass20 = planet_mass.head(20)
planet_mass20
```

```python
sns.set_theme(style='darkgrid')
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=planet_mass20,
    x="light_years_from_earth", y="orbital_radius",
    hue="discovery_date", size="planet_mass",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
